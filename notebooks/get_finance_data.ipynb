# Get finance data

```python
import yfinance as yf
import pandas as pd
import time
from datetime import timedelta
import pandas as pd
```

## 1. Get Tickets

```python
path_save = '../db/'
```

```python
sp500_wiki_url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
sp500_companies = pd.read_html(sp500_wiki_url)[0]
sp500_companies
```

```python
sp500_companies.to_csv(path_save+"sp500_wiki_database.csv", index=False, encoding="utf-8-sig")
```

```python
list_companies = sp500_companies["Symbol"].values.tolist()
list_companies
```

```python
def formato_tiempo(segundos):
    # Convertir segundos a timedelta
    td = timedelta(seconds=segundos)
    # Extraer horas, minutos y segundos
    horas, resto = divmod(td.seconds, 3600)
    minutos, segundos = divmod(resto, 60)
    # Formatear como hh:mm:ss
    return f"{horas:02d}:{minutos:02d}:{segundos:02d}"
```

```python
list_data_comp = []
list_hist_data_comp = []

s_date = '2022-10-01'
e_date = '2024-10-01'

tiempo_inicio_for =time.time()


print("Start")
for i, k in enumerate(list_companies):
    tiempo_inicio_loop = time.time()
    
    try:
        stock_info = yf.Ticker(k)

        info_ticket = stock_info.info

        #get data
        data = {
            'ticker': k,
            'country': info_ticket.get('country', None),
            'industry': info_ticket.get('industry', None),
            'sector': info_ticket.get('sector', None),
            'longBusinessSummary': info_ticket.get('longBusinessSummary', None),
            'fullTimeEmployees': info_ticket.get('fullTimeEmployees', None),
            'marketCap': info_ticket.get('marketCap', None),
            'shortName': info_ticket.get('shortName', None), 
            'companyOfficers_title': info_ticket.get('companyOfficers', [])[0].get("title", None),
            'companyOfficers_name': info_ticket.get('companyOfficers', [])[0].get("name", None)
             }


        hist_data = stock_info.history(start=s_date, end=e_date, interval="1d")
        hist_data["ticker"]=k


        list_data_comp.append(data)
        list_hist_data_comp.append(hist_data)
    
    except Exception as e:
        print(f"Error con {k}: {e}")


    tiempo_fin_loop = time.time()
    tiempo_transcurrido_loop = tiempo_fin_loop - tiempo_inicio_loop
    tiempo_transcurrido_for = tiempo_fin_loop - tiempo_inicio_for

    print(f"IteraciÃ³n {i+1}: {i/len(list_companies):.2%} Ticket:{k} \
          Tiempo loop: {formato_tiempo(tiempo_transcurrido_loop)} \
          Tiempo acum: {formato_tiempo(tiempo_transcurrido_for)}")

tiempo_fin_for =time.time()
tiempo_transcurrido_for = tiempo_fin_for - tiempo_inicio_for

print(f"Total de tiempo: {formato_tiempo(tiempo_transcurrido_for)} segundos")
```

```python
list_data_comp
df = pd.DataFrame(list_data_comp)
df
```

```python
df_hist = pd.concat(list_hist_data_comp, axis=0).reset_index()
df_hist
```
